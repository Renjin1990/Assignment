```python
import seaborn as sns
```

```python
#create a subplot contains 4 boxes of boxplots
plt.figure(figsize=(10,10))
plt.suptitle('Classification of Iris based on their PL,SL,PW and SW',fontsize=18)
plt.subplot(2,2,1)
sns.boxplot(x='Classification',y='PL',data=data)
plt.title('Classification VS Petal Length',fontsize=14,color='red')
plt.xlabel('Classification of Iris',fontsize=10)
plt.ylabel('Petal Length',fontsize=10)

plt.subplot(2,2,2)
sns.boxplot(x='Classification',y='SL',data=data)
plt.title('Classification VS Sepal Length',fontsize=14,color='red')
plt.xlabel('Classification of Iris',fontsize=10)
plt.ylabel('Sepal Length',fontsize=10)

plt.subplot(2,2,3)
sns.boxplot(x='Classification',y='SW',data=data)
plt.title('Classification VS Sepal Width',fontsize=14,color='red')
plt.xlabel('Classification of Iris',fontsize=10)
plt.ylabel('Sepal Width',fontsize=10)

plt.subplot(2,2,4)
sns.boxplot(x='Classification',y='SL',data=data)
plt.title('Classification VS Petal Width',fontsize=14,color='red')
plt.xlabel('Classification of Iris',fontsize=10)
plt.ylabel('Petal Width',fontsize=10)


plt.tight_layout()
```

```python
From this boxplot,I analysed
1.Iris-Virginica has highest Petal length,Sepal lenth and Petal Width.
2.Iris-setosa has highest sepal width,but PL,SL and PW are lowest.
3.Iris-Versicolor is more similar to iris-virginica.
```

```python
#create a heatmap of the iris dataset
sns.heatmap(data.corr(), linecolor = 'white', linewidths = 1, annot = True )
plt.title('Muti-Collinearity Heatmap for Iris dataset',fontsize=18,color='brown')
```

From the heatmap,i observed that 
    1.petal length is positively correlated with petal width,petal length and sepal length.
    2.Sepal Width is negatively correlated with petal length,petal width and sepal length


```python
#distribution of iris dataset based on the species using pairplot
sns.pairplot(data,hue="Classification")
```

```python
p=data.loc[data["Classification"] == "Iris-setosa"]
pt=p.set_index('Classification')
fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)
plt.boxplot(pt)
plt.title('Iris-Setosa',fontsize=14,color='red')
ax.set_xticklabels(['SL', 'SW',
                    'PL', 'PW'])
```
